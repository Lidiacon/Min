```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.cross_validation import cross_val_score
from sklearn import preprocessing
from sklearn.datasets import make_blobs
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
```

```python
data = pd.read_csv("adult.txt", header= None, names=['age', 'workclass', 'fnlwgt', 'education', 'education-num', 'maritalstatus',
'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
'hours-per-week', 'native-country', 'income'])
data
```

```python
data = data.replace([" ?", "?", " ?"], np.nan)
data = data.dropna()
data
```

```python
X = data[["age", "education-num", "capital-gain"]]
Y = data["income"]

forest = RandomForestClassifier(n_estimators = 50)
scores = cross_val_score(forest, X, Y, cv=5)
print("Accuracy: %0.4f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
le = preprocessing.LabelEncoder()
data["sex"] = le.fit_transform(data["sex"])
data["workclass"] = le.fit_transform(data["workclass"])
data["race"] = le.fit_transform(data["race"])
data["native-country"] = le.fit_transform(data["native-country"])
data["maritalstatus"] = le.fit_transform(data["maritalstatus"])
data["relationship"] = le.fit_transform(data["relationship"])
data
```

```python
X = data[["age", "education-num", "sex", "hours-per-week", "capital-gain", "capital-loss"]]
scores = cross_val_score(forest, X, Y, cv = 5)
print("Accuracy: %0.4f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
X = data[["workclass", "education-num", "capital-gain", "capital-loss"]]
scores = cross_val_score(forest, X, Y, cv = 5)
print("Accuracy: %0.4f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
X = data[["workclass", "education-num", "capital-gain", "capital-loss", "maritalstatus"]]
scores = cross_val_score(forest, X, Y, cv = 10)
print("Accuracy: %0.4f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
X = data[["education-num", "capital-gain", "capital-loss", "maritalstatus", "relationship", "race"]]
scores = cross_val_score(forest, X, Y, cv = 5)
print("Accuracy: %0.4f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
X = data[["education-num", "capital-gain", "capital-loss", "maritalstatus", "sex"]]
scores = cross_val_score(forest, X, Y, cv = 10)
print("Accuracy: %0.4f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
clf = DecisionTreeClassifier(max_depth= None, min_samples_split= 2)
X = data[["workclass", "education-num", "capital-gain", "capital-loss", "maritalstatus"]]
scores = cross_val_score(clf, X, Y, cv = 10)
print("Accuracy: %0.4f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
xtc = ExtraTreesClassifier(n_estimators=10, max_depth=None, min_samples_split=2)
X = data[["workclass", "education-num", "capital-gain", "capital-loss", "maritalstatus"]]
scores = cross_val_score(xtc, X, Y, cv = 10)
print("Accuracy: %0.4f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```

```python
abc = AdaBoostClassifier(n_estimators = 20)
X = data[["workclass", "education-num", "capital-gain", "capital-loss", "maritalstatus"]]
scores = cross_val_score(xtc, X, Y, cv = 10)
print("Accuracy: %0.4f (+/- %0.2f)" % (scores.mean(), scores.std() * 2))
```
